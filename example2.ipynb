```python
import pandas as pd
import matplotlib.pyplot as plt

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.barh(x_axis, y_axis,color='magenta',label='Boston')

plt.title('PyBer Fare by Month')
plt.ylabel('Date')
plt.xlabel('Fare $')
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color='black',label='New York')
ax.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan',label='Chicago')
#change order of y-axis
plt.gca().invert_yaxis()
ax.legend()
```

```python
plt.scatter(y_axis,x_axis,color='red',label='Chicago')
plt.gca().invert_yaxis()
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.legend()
```

```python
# adjustable balloon size by creating a variable that is a scaled version of y values
y_balloonsize=[]

for i in y_axis:
    y_balloonsize.append(i*5)
    
plt.scatter(y_axis,x_axis,s=y_balloonsize,color='red',label='Chicago')
plt.gca().invert_yaxis()
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.legend()
```

```python
# adjustable balloon size list comprehension built into the scatter plot function
    
plt.scatter(y_axis,x_axis,s = [i * 3 for i in y_axis],color='red',label='Chicago')
plt.gca().invert_yaxis()
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.legend()

```

```python
#now make the scatter lot using the object oriented approach

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,edgecolors='black',linewidths=1.5,s = [i * 5 for i in y_axis],color='skyblue',label='Boston')
plt.gca().invert_yaxis()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.legend()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    textprops={'color':'white'},
    autopct='%.1f%%',
        )

plt.show()
```

```python
#create the pie plot as object-oriented
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,
labels=x_axis,
textprops={'color':'white'},
autopct='%.1f%%',
explode=explode_values,
shadow=True,
counterclock=False,
startangle=95)
plt.show()
```
